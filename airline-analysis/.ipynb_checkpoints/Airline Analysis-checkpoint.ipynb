## Univariate Analysis

1. What do coach ticket prices look like? What are the high and low values? What would be considered the average? Does $500 seem like a good price for a coach ticket?

````python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math

## Read in Data
flight = pd.read_csv("flight.csv")
flight.head()
#print(flight.head())
## Task 1

sns.histplot(flight.coach_price)
plt.title("Coach Ticket Price Distribution")
plt.xlabel("Coach Ticket Price")
plt.ylabel("Count")

plt.show()
plt.clf()
print(f"Max coach price is ${flight.coach_price.max()}")
print(f"Min coach price is ${flight.coach_price.min()}")
print(f"The average price for a coach ticket is ${round(flight.coach_price.mean(), 2)}")


````

- The average price for a coach ticket was \\$376 and the min \\$44. The total max of ~120,000 flights was \\$593. Based on this information, \\$500 seems like a high price for a coach ticket.

2. Now visualize the coach ticket prices for flights that are 8 hours long. What are the high, low, and average prices for 8-hour-long flights? Does a $500 dollar ticket seem more reasonable than before?

````python
## Task 2
coach_8hours = flight.coach_price[(flight.hours == 8)]
sns.histplot(coach_8hours)
plt.title("Coach Ticket Price Distribution for 8 hour long flights")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
plt.clf()
print(f"Max coach price for 8 hour long flights is ${coach_8hours.max()}")
print(f"Min coach price for 8 hour long flights is ${coach_8hours.min()}")
print(f"The average coach ticket price for an 8 hour long flight is ${round(coach_8hours.mean(), 2)}")


````

- While prices are now closer to \\$500, it still is above average and a long way up from the minimum price of \\$170.77 for a coach ticket in an 8 hour long flight.

3. How are flight delay times distributed? Let's say there is a short amount of time between two connecting flights, and a flight delay would put the client at risk of missing their connecting flight. You want to better understand how often there are large delays so you can correctly set up connecting flights. What kinds of delays are typical?

````python
## Task 3
sns.histplot(flight.delay[flight.delay < 50], bins = 20)
plt.title("Take-off delay in minutes")
plt.xlabel("Delay (minutes)")
plt.ylabel("No. of flights")
plt.show()
plt.clf()
````

- After plotting the take-off delay for our flights, we can see there's a lot of flights that are delayed by ~10 minutes. This could be something to take into account when scheduling flights for the agency clients.

## Bivariate Analysis

4. Create a visualization that shows the relationship between coach and first-class prices. What is the relationship between these two prices? Do flights with higher coach prices always have higher first-class prices as well?

````python
## Task 4

# sns.scatterplot(data = flight, x = "coach_price", y = "firstclass_price", alpha = 0.2)
# plt.show()
# plt.clf()

# There's lots of entries, the scatter plot is very populated and can't show a lot of detail.
# Taking a sub sample of 5%

flights_sub = flight.sample(frac = 0.05)
#sns.scatterplot(data = flights_sub, x = "coach_price", y = "firstclass_price", alpha = 0.2)
sns.lmplot(data = flights_sub, x = "coach_price", y = "firstclass_price", line_kws={'color': 'black'}, lowess = True)
plt.title("Relationship between Coach and First Class ticket prices")
plt.xlabel("Coach Seat Ticket Price")
plt.ylabel("First Class Seat Ticket Price")
plt.show()
plt.clf()


````

- We find that there's a positive correlation between coach and first class ticket prices, meaning that indeed, a higher ticket price in Coach is associated with a higher ticket price in First Class.

5. What is the relationship between coach prices and inflight features &mdash; inflight meal, inflight entertainment, and inflight WiFi? Which features are associated with the highest increase in price?

````python
## Task 5
plt.figure(figsize= (10, 4))
plt.subplot(2, 2, 1)
sns.boxplot(x = "coach_price", y = "inflight_meal", data = flight, order = ['Yes', 'No'])
plt.title("Coach Price vs Included Meals")
plt.xlabel("Coach Price (in USD)")
plt.ylabel("Included Meal?")


plt.subplot(2, 2, 2)
sns.boxplot(x = "coach_price", y = "inflight_entertainment", data = flight, order = ['Yes', 'No'])
plt.title("Coach Price vs Included Entertainment")
plt.xlabel("Coach Price (in USD)")
plt.ylabel("Included Entertainment?")


plt.subplot(2, 2, 3)
sns.boxplot(x = "coach_price", y = "inflight_wifi", data = flight, order = ['Yes', 'No'])
plt.title("Coach Price vs Included Wi-Fi")
plt.xlabel("Coach Price (in USD)")
plt.ylabel("Included Wi-Fi?")

plt.tight_layout()
plt.show()
plt.clf()
````

- After plotting the price vs these different ammenities, we can see that while every included feature increases the price, some do so more than others. With Meal being the one with the least increase and Wi-Fi and Entertainment increasing it more.

6. How does the number of passengers change in relation to the length of flights?

````python
## Task 6
sns.lmplot(data = flights_sub, x = "hours", y = "passengers", x_jitter = 0.2, y_jitter = 0.2, scatter_kws={"s": 3, "alpha": 0.4}, fit_reg = False)
plt.title("No. of Passengers vs Flight Length (hours)")
plt.xlabel("Flight Length (hours)")
plt.ylabel("No. of Passengers")
plt.show()
plt.clf()
````

- From this graph, we can find that there's a lot less flights that are 6 and 8 hours long, and those that exist, usually have between 200 and 220 passengers. The rest of the flights' passenger numbers are more spread. There's also almost no flights wwith ~180 passengers.

## Multivariate Analysis

7. Visualize the relationship between coach and first-class prices on weekends compared to weekdays.

````python
## Task 7

# lm = sns.lmplot(x="total_bill", y="tip", col="smoker", data=tips)
# fig = lm.fig
# fig.suptitle("Custom Super Title")
# a1 = fig.axes[1]
# a1.set_title("Custom Title")


lm = sns.lmplot(data = flights_sub, x = "coach_price", y = "firstclass_price", line_kws = {'color': 'black', 'linewidth': 1},\
                lowess = True, col = "weekend", scatter_kws = {'alpha': 0.4, 's': 7})
fig = lm.fig
# fig.suptitle("Coach vs First Class on Weekdays & Weekends")
fig.supylabel("First Class Ticket Price")
fig.supxlabel("Coach Ticket Price")
fig.axes[0].set_title("Weekend Flights")
fig.axes[1].set_title("Weekday Flights")

lm.set_xlabels(" ")
lm.set_ylabels(" ")

plt.show()
plt.clf()
````

- It seems like the relationship between coach and first class tickets stays the same throughout all of the week. However, we also found out that overall, tickets are more expensive on the weekends vs on the weekdays.

8. How do coach prices differ for redeyes and non-redeyes on each day of the week?

````python
## Task 8
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]


sns.barplot(data = flight, x = flight["day_of_week"], y = flight.coach_price, order = day_order, hue = "redeye")
plt.legend(title = "Redeye")
plt.xlabel("Day of Week")
plt.ylabel("Coach Ticket Price")
plt.title("Coach Ticket Price for Redeye vs Non-Redeye Flights throughout the week")

plt.show()
plt.clf()
````

Thanks to this graph, we can see that:
1. Redeye flights are always cheaper by about \\$75 USD.
2. Weekend flights cost more on average than weekday flights.
